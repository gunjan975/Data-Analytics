```python
complaint_keywords = {
    "Absorption/Leak Protection": ["leak", "overflow", "wet", "stain"],
    "Comfort/Softness": ["rash", "itch", "uncomfortable", "rough"],
    "Odour/Fragrance": ["smell", "odor", "fragrance", "stinky"],
    "Eco-friendly/Natural": ["chemical", "allergy", "harsh"],
    "Design/Fit": ["tight", "loose", "wrong size", "poor fit"],
    "Trust/Hygiene": ["infection", "unsafe", "tamper", "fake"]
}
```

```python
for theme, keywords in complaint_keywords.items():
    df[theme + "_complaint"] = df[text_col].fillna("").astype(str).str.lower().apply(
        lambda x: any(kw in x for kw in keywords)
    )
```

```python
complaint_rates = df[[col for col in df.columns if "_complaint" in col]].mean() * 100
complaint_rates = complaint_rates.round(1)
```

```python
gaps = {}

for theme in themes.keys():
    claim_pct = coverage[theme] 
    complaint_col = theme + "_complaint"
    complaint_pct = complaint_rates[complaint_col]
    
    if complaint_pct > 5 and claim_pct < 20: 
        gaps[theme] = f"White space: Many complaints about {theme}, but only {claim_pct}% claim it  an undeserved need."
    elif complaint_pct > claim_pct:
        gaps[theme] = f"Credibility gap: Many complaints about {theme}, yet {claim_pct}% of products claim it  promise not believed."
```

```python
for gap in gaps.values():
    print(gap)
```

```python
products = [
    {"variant": "Standard", "pads_per_pack": 6, "target_price_per_pad": 12},
    {"variant": "Standard", "pads_per_pack": 8, "target_price_per_pad": 12},
    {"variant": "Overnight", "pads_per_pack": 6, "target_price_per_pad": 12.5},
    {"variant": "Overnight", "pads_per_pack": 8, "target_price_per_pad": 12.5}
]
```

```python
cost_per_pad = {
    "manufacturing": 5.0,   # 
    "packaging": 1.5,       # 
    "marketing_logistics": 1.0 # 
}
```

```python
for product in products:
    pads = product["pads_per_pack"]
    target_price = product["target_price_per_pad"]
    
    total_cost_per_pad = sum(cost_per_pad.values())
    total_cost_per_pack = total_cost_per_pad * pads
    retail_price_per_pack = target_price * pads
    margin_per_pack = retail_price_per_pack - total_cost_per_pack
    margin_percent = (margin_per_pack / retail_price_per_pack) * 100
    
    
    product.update({
        "total_cost_per_pad": round(total_cost_per_pad, 2),
        "total_cost_per_pack": round(total_cost_per_pack, 2),
        "retail_price_per_pack": round(retail_price_per_pack, 2),
        "margin_per_pack": round(margin_per_pack, 2),
        "margin_percent": round(margin_percent, 1)
    })
```

```python
df_plan = pd.DataFrame(products)
df_plan = df_plan[[
    "variant", "pads_per_pack", "total_cost_per_pad", "total_cost_per_pack",
    "retail_price_per_pack", "margin_per_pack", "margin_percent"
]]
```
